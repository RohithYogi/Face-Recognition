```python
from matplotlib import pyplot as plt
from numpy import linalg as LA
import random
import numpy as np
import pandas as pd
import os
import cv2
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
def train_split(l,train_percent):
    random.seed(2019)
    random.shuffle(l)
    cut = int(len(l)*train_percent)
    return l[:cut],l[cut:]
```

```python
data_path ="../AVR_data/"
folder_path = os.listdir(data_path)
folder_path = [data_path+x for x in folder_path]
train_image_names = []
test_image_names = []
count = 0
train_labels = np.array([])
test_labels = np.array([])
for path in folder_path:
    image_names = os.listdir(path)
    image_names = [path +"/"+x for x in image_names ]
    tra,tes = train_split(image_names,0.8)
    train_image_names=train_image_names + tra
    test_image_names=test_image_names +tes
    for i in tra:
        train_labels = np.append(train_labels, count)
    for i in tes:
        test_labels = np.append(test_labels, count)
    count= count+1
```

```python
train_paths,test_paths  = train_image_names , test_image_names 
image_paths = train_paths
```

```python
face_cascade = cv2.CascadeClassifier('../haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('../haarcascade_eye_tree_eyeglasses.xml')
```

```python
train_images = []
for i in image_paths:
    img = cv2.imread(i)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#     print(gray.shape)
    faces = face_cascade.detectMultiScale(gray, 1.3, 2)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
    image=roi_gray
#     print(image.shape)
# create a mask image of the same shape as input image, filled with 0s (black color)
    mask = np.zeros_like(image)
    rows, cols = mask.shape
    mask=cv2.ellipse(mask,(int(rows/2), int(cols/2)), (250,300), angle=0, startAngle=0, endAngle=360, color=(255,255,255), thickness=-1)
    result = np.bitwise_and(image,mask)
#     print(result.shape)
    result = cv2.resize(result,(256,256),interpolation = cv2.INTER_CUBIC)
    train_images.append(result.reshape((256*256)))
test_images = []
for i in test_paths:
    img = cv2.imread(i)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#     img = cv2.resize(img,(256,256),interpolation = cv2.INTER_CUBIC)
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 2)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
    image=roi_gray
# create a mask image of the same shape as input image, filled with 0s (black color)
    mask = np.zeros_like(image)
    rows, cols = mask.shape
    mask=cv2.ellipse(mask,(int(rows/2), int(cols/2)), (250,300), angle=0, startAngle=0, endAngle=360, color=(255,255,255), thickness=-1)
    result = np.bitwise_and(image,mask)
    result = cv2.resize(result,(256,256),interpolation = cv2.INTER_CUBIC)
    test_images.append(result.reshape((256*256)))
    
```

```python
train_images = np.array(train_images)
test_images = np.array(test_images)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()  
train_images = sc.fit_transform(train_images)  
test_images = sc.transform(test_images) 
```

```python
train_images.size
```

```python
lda = LDA(n_components=50)  
X_train = lda.fit_transform(train_images, train_labels)  
X_test = lda.transform(test_images)  
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(max_depth=2, random_state=0)

classifier.fit(X_train, train_labels)  
y_pred = classifier.predict(X_test) 
```

```python
from sklearn.metrics import confusion_matrix  
from sklearn.metrics import accuracy_score

cm = confusion_matrix(test_labels, y_pred)  
print(cm)  
print('Accuracy' + str(accuracy_score(test_labels, y_pred)))  
```
